```python
import download_utils
from lfw_dataset import load_lfw_dataset
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
download_utils.link_week_4_resources()
import tensorflow as tf
import keras
import numpy as np
import keras.layers as L
import keras.backend as B
from keras_utils import reset_tf_session
import cv2
import matplotlib.pyplot as plt
import keras.models as M
import keras.utils as U
```

```python
X_, attr = load_lfw_dataset(use_raw=True, dimx=32, dimy=32)

X = X_.astype('float32') / 255.0 - 0.5

X_train, X_test = train_test_split(X, test_size=0.1, random_state=42)
IMG_SHAPE = X.shape[1:]

def show_image(x):
  plt.imshow(np.clip(x + 0.5, 0, 1))

def show_grey(x):
    plt.imshow(x.reshape(x.shape[0:2]))

def show_random(count=6):
  randIndices = np.random.randint(0, X.shape[0], count)
  for i in range(randIndices.shape[0]):
    plt.subplot(2,int(count/2),i+1)
    show_image(X[randIndices[i]])
```

```python
def customLoss(X, Y, alpha=0.4):
    sobelFilter = B.variable([[[[1.,  1.]], [[0.,  2.]],[[-1.,  1.]]],
                      [[[2.,  0.]], [[0.,  0.]],[[-2.,  0.]]],
                      [[[1., -1.]], [[0., -2.]],[[-1., -1.]]]])
    def expandedSobel(inputTensor):

    #this considers data_format = 'channels_last'
        inputChannels = B.reshape(B.ones_like(inputTensor[0,0,0,:]),(1,1,-1,1))
        #if you're using 'channels_first', use inputTensor[0,:,0,0] above

        return sobelFilter * inputChannels
    def sobelLoss(yTrue,yPred):

    #get the sobel filter repeated for each input channel
        filt = expandedSobel(yTrue)

        #calculate the sobel filters for yTrue and yPred
        #this generates twice the number of input channels 
        #a X and Y channel for each input channel
        sobelTrue = B.depthwise_conv2d(yTrue,filt)
        sobelPred = B.depthwise_conv2d(yPred,filt)

        #now you just apply the mse:
        return B.mean(B.square(sobelTrue - sobelPred))
    l1 = 0
    l2 = 0
    for i in range(3):
        l1+=B.mean(B.square(X[:, :, :, i]-Y[:, :, :, i]))
    l2 = B.mean(B.square(X-Y))
    l3 = sobelLoss(X, Y)
#     L = alpha*l1 + (1-alpha)*l2 + l3
    L = l1+0.5*l2+l3
    return L
```

```python
def GSN(IMG_SHAPE, code=32):
    inputX = L.Input(shape=IMG_SHAPE)
    channels = []
    for i in range(3):
        channelIn = L.Lambda(lambda channel: B.expand_dims(channel[:, :, :, i]))(inputX)
        conv1 = L.Conv2D(filters=32, kernel_size=(3, 3), padding="same", activation="elu")(channelIn)
        maxPool1 = L.MaxPooling2D(pool_size=(2,2))(conv1)
        
        conv2 = L.Conv2D(filters=64, kernel_size=(3, 3), padding="same", activation="elu")(maxPool1)
        maxPool2 = L.MaxPooling2D(pool_size=(2,2))(conv2)
        
        conv3 = L.Conv2D(filters=128, kernel_size=(3, 3), padding="same", activation="elu")(maxPool2)
        maxPool3 = L.MaxPooling2D(pool_size=(2,2))(conv3)
        
        conv4 = L.Conv2D(filters=256, kernel_size=(3, 3), padding="same", activation="elu")(maxPool3)
        maxPool4 = L.MaxPooling2D(pool_size=(2,2))(conv4)
        
        
        flat = L.Flatten()(maxPool4)
        dense1 = L.Dense(code, activation="elu")(flat)
        
        denseR = L.Dense(2*2*32)(dense1)
        reshaped = L.Reshape((2, 2, 32))(denseR)
        convT1 = L.Conv2DTranspose(filters=128, kernel_size=(3, 3), strides=2, activation='elu', padding='same')(reshaped)
        convT2 = L.Conv2DTranspose(filters=64, kernel_size=(3, 3), strides=2, activation='elu', padding='same')(convT1)
        convT3 = L.Conv2DTranspose(filters=32, kernel_size=(3, 3), strides=2, activation='elu', padding='same')(convT2)
        convT4 = L.Conv2DTranspose(filters=1, kernel_size=(3, 3), strides=2, activation='elu', padding='same')(convT3)
        
        channels.append(convT4)
    out = L.Concatenate()(channels)
    model = M.Model(inputX, out)
    
    return model
```

```python
def build_autoencoder(image_size, code_size):
    H,W,C = image_size
    encoder = keras.models.Sequential()
    encoder.add(L.InputLayer(image_size))
    encoder.add(L.Conv2D(filters=32, kernel_size=(3, 3), padding="same", activation="elu"))
    encoder.add(L.MaxPooling2D(pool_size=(2,2)))
    encoder.add(L.Conv2D(filters=64, kernel_size=(3, 3), padding="same", activation="elu"))
    encoder.add(L.MaxPooling2D(pool_size=(2,2)))
    encoder.add(L.Conv2D(filters=128, kernel_size=(3, 3), padding="same", activation="elu"))
    encoder.add(L.MaxPooling2D(pool_size=(2,2)))
    encoder.add(L.Conv2D(filters=256, kernel_size=(3, 3), padding="same", activation="elu"))
    encoder.add(L.MaxPooling2D(pool_size=(2,2)))
    encoder.add(L.Flatten())
    encoder.add(L.Dense(code_size, activation="elu"))

    decoder = keras.models.Sequential()
    decoder.add(L.InputLayer((code_size,)))
    decoder.add(L.Dense((2*2*256),activation='elu'))#2048
    decoder.add(L.Reshape((2, 2, 256)))
    decoder.add(L.Conv2DTranspose(filters=128, kernel_size=(3, 3), strides=2, activation='elu', padding='same'))
    decoder.add(L.Conv2DTranspose(filters=64, kernel_size=(3, 3), strides=2, activation='elu', padding='same'))
    decoder.add(L.Conv2DTranspose(filters=32, kernel_size=(3, 3), strides=2, activation='elu', padding='same'))
    decoder.add(L.Conv2DTranspose(filters=3, kernel_size=(3, 3), strides=2, activation="elu", padding='same'))

    return encoder, decoder
```

```python
s = reset_tf_session()

encoder, decoder = build_autoencoder(IMG_SHAPE, code_size=96)

inp = L.Input(IMG_SHAPE)
code = encoder(inp)
reconstruction = decoder(code)

autoencoder = M.Model(inputs=inp, outputs=reconstruction)
autoencoder.compile(optimizer="adam", loss="mse")
```

```python
autoencoder.fit(x=X_train, y=X_train, epochs=30,validation_data=[X_test, X_test])
```

```python
def showImage(count, X, indices=None):
    if(indices):
        randList = np.array(indices)
        count = len(indices)
    else:
        randList = np.random.randint(0, X.shape[0], count)
    counter=1
    fig = plt.figure()
    plt.figure(figsize=(count*2, count*2))
    plt.subplot(count, 4, 1)
    plt.title("Original")
    plt.subplot(count, 4, 2)
    plt.title("Autoencoder")
    plt.subplot(count, 4, 3)
    plt.title("CS Autoencoder")
    for i in range(count):
        
    
        plt.subplot(count, 4, counter)
        plt.axis("off")
#         plt.title("Original")
        plt.imshow(X[randList[i]]+0.5)
        counter+=1
        plt.subplot(count, 4, counter)
        plt.axis("off")
#         plt.title("Autoencoder")
        plt.imshow(decoder.predict(encoder.predict(X[randList[i], :, :, :].reshape(1, 32, 32, 3))).reshape(32, 32, 3)+0.5)
        counter+=1
        plt.subplot(count, 4, counter)
        plt.axis("off")
#         plt.title("CS Autoencoder")
        plt.imshow(gsn.predict(X[randList[i]].reshape(1, 32, 32, 3)).reshape(32, 32, 3)+0.5)
        counter+=1
        plt.subplot(count, 4, counter)
        plt.axis("off")
#         plt.title("Original")
        plt.imshow(X[randList[i]]+0.5)
        counter+=1
    plt.subplots_adjust(wspace=0, hspace=0)
    plt.show()
```
